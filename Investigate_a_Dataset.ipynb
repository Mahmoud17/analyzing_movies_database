## Data Wrangling

### General Properties

```python
data = pd.read_csv("tmdb-movies.csv")
data.head()
```

### Dropping unnecessary columns

These columns won't be relevent for our analysis

```python
data.drop(['id', 'imdb_id', 'homepage', 'tagline', 'overview', 'keywords'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

### Dropping rows with no data

```python
data.dropna(inplace=True)
```

Now are data is complete and free of null values

### Dropping duplicated rows

```python
data.drop_duplicates(inplace=True)
```

### Next we will check for illogical values

```python
data.describe()
```

**We see that some movies have a budget of 0 which doesn't make sense  
also some movies have 0 revenue which is possible but highly unlikely  
So we'll drop these rows**

```python
data = data[~((data['budget'] == 0) | (data['revenue'] == 0))]
```

#### we create a profit column to help us answer questions about profitability and the success of different aspects

```python
data['profit'] = data.revenue_adj - data.budget_adj
```

**Now are data is clean**

## Exploratory Data Analysis

```python
data.hist(figsize=(15,10));
```

```python
genres_df = data.assign(genres=data['genres'].str.split('|')).explode('genres')
```

We split the genres attribute to multiple rows to facilitate analysis

### Analyzing different metrics based on genres
we will explore the total and average revenue of movies of different genres  
also the popularity and budget

### What are the most profitable  and made genres?

```python
def bar_func(series, title, x):
    series.sort_values().plot(kind='barh', figsize=(10,5))
    plt.title(title)
    plt.xlabel(x);
```

```python
bar_func(genres_df.groupby('genres').profit.sum(), title='Total profit for every genre', x='Total Profit')
```

We see that the most successful genres are Adventure, Action, Drama and Comdey

```python
bar_func(genres_df['genres'].value_counts(), title="Number of Movies of Every Genre", x='Count')
plt.ylabel("Genre");
```

Most made genre with a significant margin is drama

### How much does each genre make on average?

```python
bar_func(genres_df.groupby('genres').profit.mean(), title='Average profit for every genre', x='Average Profit')
```

We See that animation makes the most on average but as we see above the numebr of animated movies is significantly lower than most genres which could be because animated movies take much longer to make compared to live action

### What is the most popular genre?

```python
bar_func(genres_df.groupby('genres').popularity.mean(), title='Average popularity of every genre', x='Average popularity')
```

**Adventure is the most popular which explains why it makes the money, with science fiction as a very close second**

### How do different factors affect profit and popularity?

```python
def scatter(data, x, y, title, xlabel, ylabel):
    fig, ax = plt.subplots(figsize=(12,5))
    sns.regplot(ax=ax, data=data, x=x, y=y)
    plt.title(title.title())
    plt.xlabel(xlabel.title())
    plt.ylabel(ylabel.title())
    current_values = plt.gca().get_yticks()
    plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])
    current_valuesx = plt.gca().get_xticks()
    plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in current_valuesx])
```

We'll look at how different facotors correlate with profit

Does the amount of money spent on the movie correlate with a high quality movie that people like?

```python
scatter(data, 'budget_adj', 'profit', 'profit scattered versus budget', 'budget', 'profit');
```

**Some movies with extremely high budgets actually lose money!**

```python
scatter(data, 'budget_adj', 'popularity', 'popularity scattered versus budget', 'budget', 'popularity')
```

```python
scatter(data, 'budget_adj', 'vote_average', 'vote average scattered versus budget', 'budget', 'vote average')
```

We see a very weak correlation betweeen budget and profit / popularity so how much was spent on the movie doesn't tell how successful it will be.

### The relationship between runtime and profit / popularity

```python
scatter(data, 'runtime', 'profit', 'profit vs runtime', 'runtime', 'profit')
```

There is a clear congestion in the 1 to 3 hours interval where most movies fall and a decline in profit after that

```python
scatter(data, 'runtime', 'popularity', 'popularity vs runtime', 'runtime', 'popularity')
```

```python
scatter(data, 'runtime', 'vote_average', 'vote average vs runtime', 'runtime', 'vote average')
```

Most movies are within 100 to 200 minutes and as runtime gets too long we see a drop in profitability

### Who are the most commercially successful directors?

We 'll group movies by directors and sum the revenue their movies genereated adjusted for inflation

```python
data.groupby("director")['revenue_adj'].sum().to_frame().reset_index().nlargest(10, columns='revenue_adj').plot(x='director', y='revenue_adj',kind='bar', figsize=(7,5));
current_values = plt.gca().get_yticks()
plt.title("Most Successful Directors")
plt.ylabel("Revenue")
plt.legend("")
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values]);
```

Steven Spielberg is by far the most successful director with over 14 billion dollars in revenue

### What are the best selling movies of all time?

```python
top_movies = data.nlargest(10, columns='revenue_adj')[['original_title', 'revenue_adj']]
```

```python
top_movies[['original_title', 'revenue_adj']].sort_values("revenue_adj").plot(y='revenue_adj',x="original_title",kind='barh', figsize=(10,5))
plt.ylabel("Movies")
plt.xlabel("Revenue")
plt.legend("")
plt.title("Top 10 grossing movies of all time")
current_values = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in current_values]);
```
